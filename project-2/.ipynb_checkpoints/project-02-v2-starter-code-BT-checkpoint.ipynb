````python
# based on the above drugs.info() and drugs.describe(),
# due to use of age ranges under "age", it becomes an object (no issues)
# cocaine-freq, inhalant-freq, oxycontin-freq and meth-freq has invalid values, and converted into float
# data needs to be cleaned to perform eda properly
````

````python
#data cleaning

columns = ['cocaine-frequency', 'crack-frequency',
              'heroin-frequency','inhalant-frequency',
              'oxycontin-frequency','meth-frequency']

for column in columns:
    for count in range(0, len(drugs)):
        if drugs[column][count] == '-':
            drugs[column][count] = 0
    drugs[column] = drugs[column].astype(float)
````

## 8. Introduction to dealing with outliers

---

Outliers are an interesting problem in statistics, in that there is not an agreed upon best way to define them. Subjectivity in selecting and analyzing data is a problem that will recur throughout the course.

1. Pull out the rate variable from the sat dataset.
2. Are there outliers in the dataset? Define, in words, how you _numerically define outliers._
3. Print out the outliers in the dataset.
4. Remove the outliers from the dataset.
5. Compare the mean, median, and standard deviation of the "cleaned" data without outliers to the original. What is different about them and why?

````python
rate = scores['Rate'] # pull out rate column from scores, store in variable called 'rate'
````

````python
rate.describe()
````

````python
# there are outliers in the dataset.
# the 25 percentile is quite far from the mean and median.
````

````python
# defining the outliers as anything more than 1x standard deviation away from the mean

rate_upperlimit = int(rate.mean() + rate.std())
rate_lowerlimit = int(rate.mean() - rate.std())

outliers = pd.Series([x for x in rate if x not in range(rate_lowerlimit, rate_upperlimit)])
print('Outliers are: {}'.format(outliers.values))
````

````python
rate_filtered = pd.Series([x for x in rate if x not in outliers])
print('Revised list after filtering: \n{}'.format(rate_filtered.values))
````

````python
rate_filtered.describe()
````

````python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,4))
rate.hist(ax=ax1, bins=range(0,90,10), alpha=0.75)
ax1.axvline(rate.mean(), color='red', linewidth=3)
ax1.axvline(rate.mean()+rate.std(), color='green', linestyle='--', linewidth=3)
ax1.axvline(rate.mean()-rate.std(), color='green', linestyle='--', linewidth=3)

rate_filtered.hist(ax=ax2, bins=range(20,90,10), alpha=0.75)
ax2.axvline(rate_filtered.mean(), color='red', linewidth=3)
ax2.axvline(rate_filtered.mean()+rate_filtered.std(), color='green', linestyle='--', linewidth=3)
ax2.axvline(rate_filtered.mean()-rate_filtered.std(), color='green', linestyle='--', linewidth=3)
````

````python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,4))
rate.plot(kind='box', ax=ax1)

rate_filtered.plot(kind='box', ax=ax2)
````

````python
# answer:
# the outliers resulted in a much larger standard deviation i.e. 27 (before removing) vs 15 (after removing)
# after removing the outliers, we can see the 'true' median at 60, and 'true' mean at 59
````
