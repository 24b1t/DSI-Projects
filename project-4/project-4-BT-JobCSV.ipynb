```python
##############################################################
# this code is to generate the url for search page
##############################################################


def url_generator (site, title, page):
    
    # create the basic format. use with .format(job_title, page)
    url_base_careersfuture = 'https://www.mycareersfuture.sg/search?search={}&sortBy=new_posting_date&page={}'
 
    websites = {
        
        'careersfuture':
                {
                    'Data_Analyst': url_base_careersfuture.format('Data%20Analyst',page),
                    'Senior_Data_Analyst': url_base_careersfuture.format('Senior%20Data%20Analyst',page),
                    'Data_Manager': url_base_careersfuture.format('Data%20Manager',page),
                    'Data_Architect': url_base_careersfuture.format('Data%20Architect',page),
                    'Senior_Data_Architect': url_base_careersfuture.format('Senior%20Data%20Architect',page),
                    'Data_Analytics': url_base_careersfuture.format('Data%20Analytics',page),
                    'Data_Scientist': url_base_careersfuture.format('Data%20Scientist',page),
                    'Data_Science': url_base_careersfuture.format('Data%20Science',page),
                    'Business_Analytics': url_base_careersfuture.format('Business%20Analytics',page),
                    'Financial_Analytics': url_base_careersfuture.format('Financial%20Analytics',page),
                    'Marketing_Analytics': url_base_careersfuture.format('Marketing%20Analytics',page),
                    'Data_Visualisation': url_base_careersfuture.format('Data%20Visualisation',page),
                    'Data_Operations': url_base_careersfuture.format('Data%20Operations',page),
                    'Data_Strategist': url_base_careersfuture.format('Data%20Strategist',page),
                    'Data_Engineer': url_base_careersfuture.format('Data%20Engineer',page),
                    'Data_Intern': url_base_careersfuture.format('Data%20Intern',page),
                    'Data_Lead': url_base_careersfuture.format('Data%20Lead',page),
                    'Data_Consultant': url_base_careersfuture.format('Data%20Consultant',page),
                    'Machine_Learning': url_base_careersfuture.format('Machine%20Learning',page),
                    'Deep_Learning': url_base_careersfuture.format('Deep%20Learning',page),
                    'Big_Data': url_base_careersfuture.format('Big%20Data',page),
                    'Business_Intelligence': url_base_careersfuture.format('Business%20Intelligence',page),        
                
                } 
        
    }
    
    return websites[site][title]
```

```python
##############################################################
# this code is to command chromedriver to obtain html
##############################################################

def get_html (site):
    
    #driver = webdriver.Chrome(chromedriver, chrome_options = chrome_options)
    driver.get(site)
    sleep(3)
    html = driver.page_source
    #driver.close()

    return html
```

```python
##############################################################
# this code is to create a list of search criterias
##############################################################

sites = ['careersfuture']

job_titles = {
    
    'careersfuture':['Data_Analyst','Senior_Data_Analyst',
                     'Data_Manager', 'Data_Architect',
                     'Senior_Data_Architect', 'Data_Analytics',
                     'Data_Scientist', 'Data_Science',
                     'Business_Analytics', 'Financial_Analytics',
                     'Marketing_Analytics', 'Data_Visualisation',
                     'Data_Operations', 'Data_Strategist',
                     'Data_Engineer', 'Data_Intern',
                     'Data_Lead', 'Data_Consultant',
                     'Machine_Learning', 'Deep_Learning',
                     'Big_Data', 'Business_Intelligence'                    
                    ]

}
```

```python
# this list will contain all the job listings url
job_links = []
```

```python
##############################################################
# this code is to create a list of job listings URL
##############################################################

# scan through each site
for site in sites:
    
    # for careersfuture only
    if site == 'careersfuture':
        # scan through each job title
        for job_title in job_titles[site]:

            # load the first page, see how many results available

            html = get_html( url_generator(site,job_title,0))
            results_counts = Selector(text=html).xpath('//*[@id="search-results-message"]/div/text()').extract()
            
            try:
                
                num_pages = int(np.ceil(int(results_counts[0].replace(' jobs found','')) / 20.))+2
            except:
                num_pages = 10
            #print('Update: Found {} results for {} from {}'.format(results_counts[0].replace(' jobs found',''), job_title, site))

            # scan through each page
            for page in range(num_pages):

                html = get_html( url_generator(site, job_title, page))
                job_url = Selector(text=html).xpath('//*[@class="bg-white mb3 w-100 dib v-top pa3 no-underline flex-ns flex-wrap JobCard__card___22xP3"]/@href').extract()
                for item in job_url:         
                     job_links.append('https://www.mycareersfuture.sg' + item)
                print('Update: Added {} entries for page {} of from {}-{}'.format(len(job_url), page, site, job_title))

print('Total of {} job entries found.'.format(len(job_links)))    
```

```python
# create a list for collecting all job listing details
job_details = pd.DataFrame()
job_pull_error_list = []
job_pull_error_list_2 = []
```

```python
##############################################################
# this function is to get join details from careerfuture url
# and return the job details
##############################################################

def pull_job_details_mycareersfuture (url):
    
    # load page, get page source, close page
    #driver = webdriver.Chrome(chromedriver, chrome_options = chrome_options)
    html = get_html(url)
    
    # get the job details
    job_id = Selector(text=html).xpath('//*[@data-cy="jobinfo__jobpostid--span"]/text()').extract()[0].replace(" ",'')
    job_title = Selector(text=html).xpath('//*[@id="job_title"]/text()').extract()[0]

    company_name = Selector(text=html).xpath('//*[@name="company"]/text()').extract()[0]
    # cater for cases where company name is undisclosed
    try:
        company_address = Selector(text=html).xpath('//*[@href="#location_map"]/text()').extract()[0]
    except:
        company_address = np.nan

        
    employment_type = Selector(text=html).xpath('//*[@id="employment_type"]/text()').extract()[0]
        
    seniority = Selector(text=html).xpath('//*[@id="seniority"]/text()').extract()[0]

    job_category = Selector(text=html).xpath('//*[@id="job-categories"]/text()').extract()[0]
    job_url = url
    job_description = Selector(text=html).xpath('//*[@id="description-content"]/p/text()').extract()
    job_requirements = Selector(text=html).xpath('//*[@id="requirements-content"]/ul/li/text()').extract()
    
    job_skills = Selector(text=html).xpath('//*[@id="skills-needed"]/div/div/label/text()').extract()

    # salary could be empty. use try/except
    try:
        pay_min = Selector(text=html).xpath('//*[@class="dib"]/text()').extract()[0]
        pay_max = Selector(text=html).xpath('//*[@class="dib"]/text()').extract()[1]
        pay_type = Selector(text=html).xpath('//*[@class="salary_type dib f5 fw4 black-60 pr1 i pb"]/text()').extract()[0]
    except:
        pay_min = np.nan
        pay_max = np.nan
        pay_type = np.nan

    post_date = Selector(text=html).xpath('//*[@id="last_posted_date"]/text()').extract()[0].replace('Posted ','')
    close_date = Selector(text=html).xpath('//*[@class="purple f6 fw4 pr3 pt2 dib"]/text()').extract()[0].replace('Closing on ','')

    return [job_id, job_title, company_name, company_address,  
                                      job_category, job_url, job_description, job_requirements,
                                      employment_type, seniority, job_skills,
                                      pay_min, pay_max, pay_type, post_date, close_date]
```

```python
##############################################################
# this function is to get join details from the URL provided
# and return the job details
##############################################################

def pull_job_details(url, web):
    
    if web == 'mcf':
        return pull_job_details_mycareersfuture(url)
    
```

```python
##############################################################
# populate the 'job_details' dataframe with url in 'job_links'
##############################################################

for row_count in range(len(job_links)):
    print('Update: Pulling data for job {}'.format(row_count))
    try:
        job_details = job_details.append([pull_job_details(job_links[row_count], 'mcf')])
    except:
        print('Error: Error pulling data for job {}, executing attempt 2'.format(row_count))
        try:
            job_details = job_details.append([pull_job_details(job_links[row_count], 'mcf')])
        except:
            print('Error: Error pulling data for job {}, executing attempt 3'.format(row_count))
            try:
                job_details = job_details.append([pull_job_details(job_links[row_count], 'mcf')])
            except:
                print('Error: Error pulling data for job {}, ignoring row'.format(row_count))
                job_pull_error_list.append(row_count)
```

```python
##############################################################
# identify which job_pull failed
##############################################################

print('Failed to retrieve {} items from site.\nIndex: {}'.format(
     len(job_pull_error_list), job_pull_error_list))
```

```python
##############################################################
# identify which job_pull failed second attempt - data ignored
##############################################################

print('Failed to retrieve {} items from site.\nIndex: {}'.format(
     len(job_pull_error_list_2), job_pull_error_list_2))
```

```python
##############################################################
# all web-jobs complete. closing driver.
##############################################################

driver.close()
```

```python
##############################################################
# Data Cleaning
##############################################################

# reset the index of the dataframe
job_details.reset_index(drop=True, inplace=True)

# add back the column names 
job_details.columns = ['job_id', 'job_title', 'company_name', 'company_address', 
                       'job_category', 'job_url', 'job_description', 'job_requirements', 
                       'employment_type', 'seniority', 'job_skills',
                       'pay_min', 'pay_max', 'pay_type', 'posting_date', 'closing_date'
                       ]

job_count = len(job_details)

# remove entries from duplicated job_id (unique identifier)
job_details.drop_duplicates(subset='job_id', keep='first', inplace=True)
job_details.reset_index(drop=True, inplace=True)

print('{} duplicate entries were removed. \nCurrent Dataset contains {} entries'.format(\
      job_count - len(job_details), len(job_details)))
```

```python
##############################################################
# Do some basic data formatting
##############################################################


output = job_details.copy()
output['job_requirements'] = output['job_requirements'].map(lambda x: ' '.join(x))
output['job_description'] = output['job_description'].map(lambda x: ' '.join(x).replace('\xa0',''))
```

```python
##############################################################
# export data into csv file
##############################################################

output.to_csv('./job_openings.csv')
```
